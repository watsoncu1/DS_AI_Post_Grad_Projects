```python
### Python Code For Project: Self-Pay Kiosk ###


class SelfPayKiosk:
    SALES_TAX = 0.07

    def __init__(self):
        self.num_customers = 0
        self.total_sales = 0.0
        self.amount_due = 0.0
        self.checked_out = False

    def get_total_sales(self):
        return self.total_sales

    def get_amount_due(self):
        return self.amount_due

    def get_num_customers(self):
        return self.num_customers

    def scan_item(self, price):
        if price > 0:
            self.amount_due += price
        else:
            print("Invalid price. Item not added.")

    def check_out(self):
        if self.amount_due > 0:
            self.amount_due *= (1 + self.SALES_TAX)
            self.checked_out = True
        else:
            print("No items to check out.")

    def make_payment(self, payment):
        if self.checked_out and payment > 0:
            if payment >= self.amount_due:
                self.total_sales += self.amount_due
                self.num_customers += 1
                self.amount_due = 0.0
                self.checked_out = False
                print("Transaction completed. Thank you!")
            else:
                self.amount_due -= payment
                print(f"Partial payment made. {self.amount_due:.2f} remaining.")
        else:
            print("Payment cannot be processed at this time.")

    def reset_kiosk(self):
        self.__init__()

    def cancel_transaction(self):
        if not self.checked_out:
            self.amount_due = 0.0
            print("Transaction cancelled.")
        else:
            print("Cannot cancel after checkout. Please complete payment.")

    def simulate_sales(self, num_sales, initial_price, incr_price):
        for _ in range(num_sales):
            self.scan_item(initial_price)
            self.check_out()
            self.make_payment(self.amount_due + 1)
            initial_price += incr_price

if __name__ == '__main__':
    kiosk = SelfPayKiosk()
    # Demonstrates enhanced user experience with clear feedback and validation.


```

## **2. Self-Initiated Exploration in Programming Description:**

#### The Self-Pay Kiosk program is a Python-based simulation designed to emulate the transactional process of a self-service checkout system commonly found in retail environments. This program encapsulates several fundamental concepts of software development and design, specifically focusing on object-oriented programming (OOP), user input handling, and conditional logic. The explanation below outlines the operation of the program, detailing its structure, functionality, and the programming concepts it employs, intended for an admission committee's review for an Online Master of Science in Data Science degree program.

```python
class SelfPayKiosk:
    SALES_TAX = 0.07

    def __init__(self):
        self.num_customers = 0
        self.total_sales = 0.0
        self.amount_due = 0.0
        self.items_prices = []
        print("\nWelcome to the Self-Pay Kiosk! Let's begin your transaction.\n")

    def prompt_for_price(self):
        while True:
            response = input("Enter item price or type 'CT' for checkout ($XX.XX or 'CT'): ").strip().lower()
            if response in ['exit', 'escape']:
                print("Exiting the kiosk. Thank you for visiting.")
                exit()
            elif response == 'ct':
                return None  # Checkout command issued
            else:
                try:
                    price = float(response)
                    if price > 0:
                        print(f"Item price entered: ${price:.2f}")
                        return price
                    else:
                        print("Please enter a positive number for the item price.")
                except ValueError:
                    print("Invalid input. Please enter a valid number or type 'CT' to proceed to checkout.")

    def scan_item(self):
        print("Please scan your items. Enter the price or type 'CT' when you are ready to checkout.")
        while True:
            price = self.prompt_for_price()
            if price is None:  # User ready to checkout
                break
            self.items_prices.append(price)  # Add the valid item price to the list
            self.amount_due += price
            print(f"Current total: ${self.amount_due:.2f}")

    def check_out(self):
        if self.amount_due > 0:
            # Apply sales tax to the total amount due
            self.amount_due *= (1 + SelfPayKiosk.SALES_TAX)
            print(f"\nTotal amount due (including tax): ${self.amount_due:.2f}")
        else:
            print("No items scanned. Exiting the kiosk.")
            exit()

    def process_payment(self):
        while self.amount_due > 0:
            try:
                payment = float(input("Enter payment amount: $"))
                if payment <= 0:
                    print("Please enter a positive amount for the payment.")
                    continue
                print(f"Payment entered: ${payment:.2f}")
                self.amount_due -= payment
                if self.amount_due > 0:
                    print(f"Remaining amount due: ${self.amount_due:.2f}")
                elif self.amount_due < 0:
                    print(f"Change due to customer: ${-self.amount_due:.2f}")
                    self.amount_due = 0  # Reset the amount due to exit the loop
                else:
                    print("Payment complete. No change due.")
            except ValueError:
                print("Invalid input. Please enter a numeric value for the payment.")

    def run(self):
        self.scan_item()
        self.check_out()
        self.process_payment()
        self.num_customers += 1
        print(f"Transaction complete. Total customers served today: {self.num_customers}.\n")

if __name__ == "__main__":
    kiosk = SelfPayKiosk()
    kiosk.run()
```
