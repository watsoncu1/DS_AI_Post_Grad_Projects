### 3. What are the attributes that have a strong correlation with the target attribute (personal loan)?

```python
# Calculating the correlation matrix to examine relationships between attributes.
# A correlation matrix is a table showing correlation coefficients between variables. Each cell in the table shows the correlation
# between two variables. The value is in the range of -1 to 1 where 1 indicates a perfect positive correlation, -1 indicates a perfect
# negative correlation, and 0 indicates no correlation at all. This is a crucial step in exploratory data analysis (EDA) as it helps
# in identifying potential relationships or dependencies between variables, which can inform feature selection and model building in
# machine learning projects. It provides a quantitative measure of how changes in one variable are associated with changes in another.
correlation_matrix = data.corr() # Output of correlation_matrix for analysis and interpretation and visualization.

# Extracting correlations of features with the 'Personal_Loan' attribute and sorting in descending order.
# This operation focuses on understanding how each attribute in the dataset correlates with the 'Personal_Loan' attribute specifically.
# Sorting these correlation coefficients in descending order allows us to quickly identify which features have the strongest positive
# or negative associations with the likelihood of a personal loan, which can be insightful for identifying key factors that influence
# a customer's decision to take a loan. Such insights are valuable for feature engineering and for designing targeted strategies in
# marketing or risk assessment.
personal_loan_correlation = correlation_matrix['Personal_Loan'].sort_values(ascending=False) # Output of personal_loan_correlation for analysis and interpretation.

# Displaying the sorted correlations.
# By outputting the sorted list of correlation coefficients with 'Personal_Loan', we can easily review and interpret the relationships.
# For accessibility, when discussing or sharing these findings, especially in a presentation or report, it's beneficial to describe
# the insights in text or through accessible data visualization techniques. This ensures that the information is conveyed effectively
# to all audience members, including those with visual impairments, enhancing the inclusivity of the analysis.
personal_loan_correlation # Output of personal_loan_correlation for analysis and interpretation.
```

1. Income: There is a **moderate positive correlation** (correlation coefficient of 0.502) between income levels and the likelihood of accepting a personal loan. Higher income is associated with greater interest in personal loans.

2. CCAvg (Credit Card Spending): Customers who spend more on credit cards per month exhibit a moderate positive correlation (correlation coefficient of 0.367) with personal loan acceptance.

3. CD_Account (Certificate of Deposit): Holding a certificate of deposit (CD) account with the bank shows a **moderate positive correlation** (correlation coefficient of 0.316) with accepting personal loans.

4. Mortgage: Although weaker, there is a positive correlation (correlation coefficient of 0.142) between higher mortgage values and personal loan acceptance.

5. Education: Slightly higher education levels are associated with a small positive correlation (correlation coefficient of 0.137) regarding personal loan interest.

In conclusion, financial stability (income, credit card spending, and CD accounts) plays a significant role in a customer's inclination toward personal loans.

### Integrated Boxplot and Histogram Visualization Function for Data Analysis





```python
def histogram_boxplot(data, feature, figsize=(12, 7), kde=False, bins=None): # Define the function with the required parameters.
    """
    Visualizes the distribution of a numerical feature with a histogram and boxplot on the same x-axis.

    Parameters:
    - data (DataFrame): Dataset containing the features.
    - feature (str): Numerical feature to visualize.
    - figsize (tuple): Size of the figure (width, height).
    - kde (bool): Overlay a Kernel Density Estimate (KDE) for a smooth distribution.
    - bins (int/sequence/str): Define the number of bins for the histogram.

    Returns:
    - None: Displays the plot.
    """

    # Creating subplots for a cohesive visual comparison of two different plot types.
    # The shared x-axis allows for direct comparison of the feature's distribution in both plot styles.
    f2, (ax_box2, ax_hist2) = plt.subplots( # Two subplots, one for the boxplot and one for the histogram.
        nrows=2,  # Two rows: one for the boxplot, one for the histogram.
        sharex=True,  # Shared x-axis for aligned comparison.
        gridspec_kw={"height_ratios": (0.25, 0.75)},  # More space allocated to the histogram for detail.
        figsize=figsize,  # Custom figure size for adequate space and visibility.
    )

    # Drawing a boxplot to examine the feature's quartiles, median, and potential outliers.
    # The color is chosen for contrast and visibility, important for understanding the distribution's summary.
    sns.boxplot(
        data=data, x=feature, ax=ax_box2, showmeans=True, color="#800080"  # Purple is used for the boxplot for an aesthetically pleasing contrast and visibility.
    )  # Purple is used for the boxplot for an aesthetically pleasing contrast and visibility.

    # Setting up the histogram to visualize the feature's distribution.
    # The 'magma' color palette is used here for the histogram for its visual appeal and clarity.
    # If 'bins' is specified, it's used to customize the granularity of the histogram; otherwise, the default is used.
    if bins:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, bins=bins, palette="magma") # 'magma' palette for clear, distinct visualization.
    else:
        sns.histplot(data=data, x=feature, kde=kde, ax=ax_hist2, palette="magma")  # 'magma' palette for clear, distinct visualization.

    # Highlighting the mean and median on the histogram with vertical lines for added context.
    # Green is used for the mean and black for the median to stand out against the histogram's colors.
    ax_hist2.axvline( # Highlighting the mean on the histogram for readability and contrast.
        data[feature].mean(), color="#008000", linestyle="--", label='Mean' # Green line for the mean to provide a visual contrast with the histogram's colors.
    )  # Green line for the mean provides a visual contrast.
    ax_hist2.axvline( # Highlighting the median on the histogram for readability.
        data[feature].median(), color="#000000", linestyle="-", label='Median' # Black line for the median to stand out against the histogram's colors.
    )  # Black line for the median ensures visibility.
    ax_hist2.legend()  # Display a legend to label the mean and median lines.

# Using this function will give a comprehensive view of the feature's distribution,
# showing both the detailed spread of data points and a summary through the boxplot.
# Such combined visualizations offer a richer understanding than either plot would alone.
```

The above code for a histogram_boxplot function integrates histograms and boxplots, providing a comprehensive analysis of data distributions with minimal input. It enhances interpretability by incorporating key statistical metrics like mean and median on the visualizations. ### Enhanced Barplot Visualization with Optional Percentage Labels

```python
def labeled_barplot(data, feature, perc=False, n=None):# Creating a labeled barplot for categorical features for more intiutive analysis
    """
    Creates a barplot for a categorical variable with value counts or percentages labeled on each bar.

    Parameters:
    - data (DataFrame): The dataset containing the feature of interest.
    - feature (str): The name of the categorical feature to be visualized.
    - perc (bool, optional): Determines whether to show percentages instead of counts. Default is False.
    - n (int, optional): The number of top categories to display, based on frequency. Displays all if None.

    Returns:
    - None: The plot is displayed and nothing is returned.
    """

    # Calculating the total number of occurrences for the feature to facilitate percentage calculation if required.
    total = len(data[feature])  # Total instances within the feature column.

    # Dynamically determining the figure size based on the number of unique categories in the feature.
    # This ensures the plot is sized appropriately to accommodate all categories without crowding.
    count = data[feature].nunique()  # The count of unique categories within the feature.
    plt.figure(figsize=(n + 1, 5) if n else (count + 1, 5))  # Set the figure size dynamically.

    # Customizing the appearance of the plot for clear and accessible visualization.
    plt.xticks(rotation=90, fontsize=15)  # Rotate x-axis labels for better readability, especially for longer category names.

    # Constructing the count plot using Seaborn's visually distinct 'Set2' palette for clear differentiation of categories.
    # The palette choice is particularly important to accommodate individuals with visual impairments by providing distinct, contrasting colors.
    ax = sns.countplot( # Create the count plot with Seaborn.
        data=data, # The dataset containing the feature of interest.
        x=feature, # The categorical feature to be visualized.
        palette="Set2",  # The 'Set2' palette is chosen for clear color distinction among categories.
        order=data[feature].value_counts().index[:n].sort_values()  # Sorting categories by frequency for a meaningful display order.
    )

    # Annotating each bar with the count or percentage value to convey exact quantities.
    # Providing numerical annotations on the bars makes the plot information accessible even without the ability to distinguish colors.
    for p in ax.patches: # Iterate over each bar in the plot.
        # Format the label as a percentage if 'perc' is True; otherwise, use the count.
        label = "{:.1f}%".format(100 * p.get_height() / total) if perc else int(p.get_height()) # Format the label as a percentage or count.

        # Determine the position for each label to ensure it's centered and legible above its respective bar.
        ax.annotate( # Annotate the text on each bar.
            label, # Label text for each bar.
            (p.get_x() + p.get_width() / 2, p.get_height()),  # Position the text above the middle of the bar.
            ha='center',  # Horizontally center the label.
            va='center',  # Vertically center the label just above the bar.
            size=12,  # Text size is chosen to be readable yet not overwhelm the visual.
            xytext=(0, 5),  # Offset the text by 5 points to prevent overlap with the bar.
            textcoords="offset points"  # Use offset points for positioning the text.
        )

    # Display the final plot which now includes labels on each bar for a precise and comprehensive visualization.
    plt.show() # Display the plot to the user.
```

The above code defines a Python function, labeled_barplot, which generates barplots with an option to display percentages or counts of categories within a specified feature of a DataFrame. This function is tailored to enhance data visualization by allowing the display of either the top n categories or all, depending on the user's choice, thereby offering flexibility in data presentation. By incorporating adjustable sizing, rotation for readability, and a paired color palette for aesthetic appeal, the function showcases practical data visualization techniques. Such a tool highlights an applicant's ability to create insightful and customizable visualizations, a valuable skill in data science and artificial intelligence disciplines.

**Note:** **Kurtosis** is a statistical measure that describes the **"tailedness"** of the probability distribution of a real-valued random variable. It provides insights into how much of a distribution's data is concentrated in the tails (extreme values) relative to the center of the distribution (the mean). Here are some key points about kurtosis:

1. Tails and Outliers:
   - The tails of a distribution represent the probability or frequency of values that are **extremely high or low** compared to the mean.
   - Kurtosis helps us understand how often **outliers** occur in a dataset.

2. Types of Kurtosis:
   - Distributions can be categorized based on their kurtosis:
     - Mesokurtic: Medium-tailed distributions where outliers are neither highly frequent nor highly infrequent. These distributions have a kurtosis close to 3 (or an excess kurtosis close to 0).
     - Platykurtic: Thin-tailed distributions with low kurtosis. They have fewer outliers.
     - Leptokurtic: Fat-tailed distributions with high kurtosis. These distributions have more frequent outliers.

3. Comparison to Normal Distribution:
   - Normal distributions (bell-shaped curves) have a kurtosis of exactly 3.
   - Excess kurtosis (kurtosis minus 3) makes it easier to compare other distributions to the normal distribution:
     - Excess kurtosis of 0 corresponds to a mesokurtic distribution.

4. Interpretation:
   - Understanding kurtosis helps in analyzing regional variations, preferences for banking products, income levels, and loan acceptance rates.

In summary, kurtosis provides valuable information about the shape and behavior of data distributions, aiding decision-making in fields like data science and artificial intelligence¹²³⁴. 📊🌟

```python
def labeled_barplot(data, feature, perc=False, n=None):# Define the function.
    """
    Generates a labeled barplot for a categorical feature within a dataset. This visualization is instrumental in comparing
    the frequency or proportion of categories visually. It's beneficial for preliminary data analysis, allowing for a quick
    assessment of the feature's distribution and prominence of its categories.

    Parameters:
    - data (DataFrame): The dataset that contains the feature of interest.
    - feature (str): The name of the categorical feature to be visualized.
    - perc (bool, optional): If True, the bar labels display percentages instead of raw counts. Useful for relative comparison.
    - n (int, optional): Limits the number of displayed categories to the 'n' most frequent ones. By default, all categories are shown.

    Returns:
    - None: The function plots the barplot and does not return any value.

    Note: This visualization has been designed to accommodate individuals with visual impairments, using a color palette with distinct colors.
    """

    # Calculate total entries to convert counts to percentages if required.
    total = len(data[feature]) # Count the total number of entries in the feature.
    # Determine the count of unique categories to adjust the plot size dynamically, ensuring all bars are visible and legible.
    count = data[feature].nunique() # Count the number of unique categories.

    # Setting the figure size to accommodate the number of categories.
    # This ensures that the plot is sufficiently large for readability, particularly important for accessibility considerations.
    plt.figure(figsize=(n + 1, 5) if n else (count + 1, 5)) # Adjust the size of the plot based on the number of categories.

    # Enhancing the readability of x-axis labels with adjustments, beneficial for all viewers, including those with visual impairments.
    plt.xticks(rotation=90, fontsize=15) # Rotate the x-axis labels by 90 degrees and adjust the font size.

    # Creating the countplot with a color palette that is easily distinguishable, aiding viewers with color vision deficiencies.
    ax = sns.countplot( # Plotting the count of each category.
        data=data, # Providing the data
        x=feature, # The categorical feature to be visualized.
        palette="Paired",  # "Paired" palette is selected to accommodate individuals with visual impairments with distinct colors.
        order=data[feature].value_counts().index[:n].sort_values()  # Sorting categories by frequency to prioritize significant ones.
    )

    # Annotating each bar with its count or percentage, improving data interpretation.
    for p in ax.patches:# Loop through each bar in the plot.
        label = "{:.1f}%".format(100 * p.get_height() / total) if perc else p.get_height() # Convert counts to percentages if required.

        # Positioning labels centrally above each bar enhances clarity and accessibility.
        ax.annotate( # Annotate each bar with its count or percentage.
            label, # Label text.
            (p.get_x() + p.get_width() / 2, p.get_height()),
            ha="center",  # Horizontal alignment.
            va="center",  # Vertical alignment, adjusted for clear visibility.
            size=12,  # Font size selected for ease of reading.
            xytext=(0, 5),  # Offset to avoid overlapping with the bar.
            textcoords="offset points",  # Offset type.
        )

    # Display the plot. This barplot serves as a foundational tool in EDA, offering insights into the distribution of categorical data.
    plt.show() # Display the plot.

# Visualizing the 'Family' feature as an example, showcasing the function's application.
labeled_barplot(data, "Family", perc=True) # Change the 'perc' parameter to True to display counts instead of percentages.
```

Reference Notes:

1. Synthetic Minority Oversampling Technique (SMOTE):
   - SMOTE is a powerful method used to address the class imbalance problem in machine learning.
   - Class imbalance occurs when one class (the minority class) is under-represented in the dataset compared to another class (the majority class).
   - Objective: To create synthetic examples of the minority class to balance the class distribution.
   - How SMOTE Works:
     - For each minority class instance, SMOTE generates new synthetic instances by interpolating between that instance and its K nearest neighbors.
     - The synthetic instances lie on the line segment connecting the original instance and its neighbors.
     - By doing this, SMOTE increases the representation of the minority class.
   - Advantages:
     - Helps improve model performance by providing more balanced training data.
     - Reduces the risk of the model being biased toward the majority class.
   - Limitations:
     - SMOTE assumes that the decision boundary is linear, which may not hold for all datasets.
     - Generated synthetic samples may not perfectly represent the true underlying distribution.
   - Application: SMOTE is widely used in fraud detection, medical diagnosis, and other domains where class imbalance is common¹².

2. Cost-Sensitive Learning:
   - Objective: To address class imbalance by assigning different costs (weights) to different misclassification errors.
   - How It Works:
     - Assigns higher misclassification costs to the minority class (which is more critical to detect correctly).
     - Adjusts the decision boundary to minimize the overall cost, considering both false positives and false negatives.
   - Advantages:
     - Allows the model to focus on minimizing costly errors.
     - Can be used with various classifiers (e.g., SVM, decision trees, neural networks).
   - Limitations:
     - Requires knowledge of the relative costs of different misclassifications.
     - May not work well if the cost matrix is not accurately specified.
   - Application: Cost-sensitive learning is useful in scenarios like medical diagnosis (where false negatives are costly) or credit card fraud detection (where false positives are costly)²³.

In summary, both SMOTE and cost-sensitive learning are essential techniques for handling class imbalance and improving model performance in real-world applications. 🌟🤖

Reference Notes:The **Area Under the Precision-Recall Curve (AUPRC)** is a valuable metric used in evaluating binary classification models, especially when dealing with imbalanced datasets. Let's break it down:

1. Precision-Recall Curve:
   - The Precision-Recall (PR) curve is a graphical representation that shows the trade-off between precision and recall (sensitivity) for different classification thresholds.
   - It is particularly useful when one class (usually the minority class) is rare or under-represented in the dataset.

2. Precision and Recall:
   - Precision: The proportion of true positive predictions (correctly predicted positive instances) out of all positive predictions (true positives + false positives).
   - Recall: The proportion of true positive predictions out of all actual positive instances (true positives + false negatives).

3. AUPRC Calculation:
   - The AUPRC summarizes the PR curve by calculating the area under it.
   - It ranges from 0 to 1, where:
     - A score of 1 indicates perfect precision-recall trade-off.
     - A score of 0 indicates poor performance.
   - A higher AUPRC suggests better model performance in capturing positive instances while maintaining high precision.

4. Use Cases:
   - AUPRC is commonly used in fields such as medical diagnosis (where false negatives are critical) and fraud detection (where false positives are costly).
   - It provides insights into how well a model balances precision and recall across different decision thresholds.

In summary, AUPRC complements other evaluation metrics (such as ROC AUC) by focusing on precision and recall, making it particularly relevant for imbalanced classification problems¹². 📊🌟

### Stacked Barplot Function for Comparative Analysis of Predictor and Target Variables:

```python

def stacked_barplot(data, predictor, target):
    """
    Generate a stacked barplot for comparing the distribution of a categorical target variable across different categories of a predictor variable.

    This function calculates the distribution of categories in the target variable within each category of the predictor variable. It then visualizes these distributions using a stacked barplot, which is particularly useful for identifying patterns or disparities across categories.

    Parameters:
    - data (DataFrame): The DataFrame containing the dataset.
    - predictor (str): The name of the predictor feature/column.
    - target (str): The name of the target feature/column.

    Returns:
    - None: This function displays the stacked barplot.
    """
    # Counting unique categories in the predictor variable to determine the width of the plot dynamically.
    count = data[predictor].nunique()

    # Sorting the target variable by frequency to establish a consistent order in the visualization.
    sorter = data[target].value_counts().index[-1]

    # Creating a cross-tabulation to show the frequency of each target category across predictor categories, sorted for visual clarity.
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(by=sorter, ascending=False)
    print(tab1)
    print("-" * 120)  # Separator line for clarity in output

    # Normalizing the table by row to compare proportional distributions across different predictor categories, enhancing comparability.
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(by=sorter, ascending=False)

    # Plotting the normalized data as a stacked barplot, where each bar's segments represent the proportion of each target category within a predictor category.
    ax = tab.plot(kind="bar", stacked=True, figsize=(count + 5, 5), color=sns.color_palette("Set2"))  # Using 'Set2' for distinct, high-contrast colors.

    # Enhancing plot annotations for accessibility:
    for bars in ax.containers:
        ax.bar_label(bars, label_type='center', fmt='%.2f%%', weight='bold')  # Bold percentages for clarity.

    plt.legend(loc="upper left", bbox_to_anchor=(1, 1), frameon=False)  # Adjusting legend position and style for accessibility.

    plt.show()

# Note: Ensure 'data', 'predictor', and 'target' are defined in your context before calling this function.
# Example usage:
# stacked_barplot(data, "Education", "Personal_Loan")
```

The above stacked_barplot function first generates a contingency table, which is essentially a tabular representation of the frequency distribution of the variables. This table, referred to as tab1 in the code, displays the count of occurrences for each combination of the predictor and target variable categories. It's sorted by the frequency of the least common target class to highlight the distribution patterns that may be less evident due to smaller sample sizes.

Additionally, the function calculates a normalized version of this table, labeled tab, which expresses the data as a percentage over the total counts for each category of the predictor variable. This normalization process converts the raw counts into a relative measure, facilitating comparisons across different categories that may have varied total counts.

Reference Notes: Certainly! Let's explore both concepts:

1. Information Gain:
   - Information gain is a concept used in decision trees and machine learning.
   - It measures how much useful information a particular feature provides in predicting the target variable.
   - Specifically, it quantifies the reduction in uncertainty (or entropy) about the target variable after considering a specific feature.
   - Formula:
     - The information gain for a feature is calculated as the difference between the entropy of the entire dataset and the weighted average entropy of the dataset after splitting it based on that feature.
   - Application:
     - Decision trees use information gain to decide which feature to split on at each node, aiming to maximize predictive accuracy.

#### 2. Chi-Squared Tests:
   - Chi-squared tests are statistical hypothesis tests used to analyze categorical data.
   - They compare the observed frequencies of different categories with the expected frequencies (based on a null hypothesis).
   - Types:
     - Chi-Square Goodness of Fit Test: Compares observed frequencies with expected frequencies for a single categorical variable.
     - Chi-Square Test of Independence: Tests whether two categorical variables are related to each other.
   - Application:
     - Used in social sciences, biology, in this case marketing, and more to assess associations and independence between categorical variables.

In summary, information gain helps decision trees choose the most informative features, while chi-squared tests evaluate relationships between categorical variables¹⁴⁵. 🌟📊

### Quantitative Outlier Detection through Interquartile Range Analysis:

```python
# Calculate the 1st quartile (Q1), where 25% of the data falls below this value
Q1 = np.percentile(data, 25)

# Calculate the 3rd quartile (Q3), where 75% of the data falls below this value
Q3 = np.percentile(data, 75)

# Calculate the Interquartile Range (IQR), representing the middle 50% of the data's spread
IQR = Q3 - Q1

# Define the lower outlier threshold: 1.5 * IQR below the 1st quartile (Q1)
lower = Q1 - 1.5 * IQR

# Define the upper outlier threshold: 1.5 * IQR above the 3rd quartile (Q3)
upper = Q3 + 1.5 * IQR
```

### Observations:
The above code executes a quantitative analysis using percentile-based statistical methods to calculate the Interquartile Range (IQR), delineating the middle 50% of the dataset between the 25th (Q1) and 75th (Q3) percentiles. The IQR is a central measure of variability that is particularly robust against the influence of outliers. The formulation of the lower and upper bounds, computed as 1.5 times the IQR below Q1 and above Q3 respectively, establishes the criteria for outlier demarcation within the dataset.

These computed boundaries are quintessential for rigorous outlier identification in exploratory data analysis, serving as a criterion for the assessment of data point normality within the statistical distribution framework. ### Implementation of a Multimetric Evaluation Framework for Scikit-learn Classification Models:

```python
# defining a function to compute different metrics to check performance of a classification model
def model_performance_classification_sklearn(model, predictors, target):# predictors and target
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    acc = accuracy_score(target, pred)  # Calculating accuracy of the model for each class
    recall = recall_score(target, pred)  # Calculating recall of the model for each class
    precision = precision_score(target, pred)  # Calculating precision of the model for each class
    f1 = f1_score(target, pred)  # Calculating f1 score of the model for each class

    # Constructing a dataframe of metrics
    df_perf = pd.DataFrame( # dataframe of metrics
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,}, # index=[0]
        index=[0], # index=[0]
    ) # index=[0]

    return df_perf# returning the dataframe
```

### Observations:
The above function, model_performance_classification_sklearn, operationalizes an advanced multimetric evaluation framework for analysis of the efficacy of classification models constructed utlizing Scikit-learn. By performing predictive analysis on a set of independent variables, the above function harnesses the model's inferential capability to generate predictions, which are subsequently juxtaposed against actual target outcomes to derive a spectrum of performance metrics.

### Metrics:
- #### **Accuracy** metric quantitatively captures the model's global prediction fidelity, reflecting the quotient of accurately predicted instances over the aggregate sample space.
- #### **Recall**, also denoted as sensitivity, probes the model's proficiency in exhaustively retrieving pertinent positive instances, essentially quantifying the fraction of true positives effectively captured from the total actual positives.
- #### **Precision** metric evaluates the exactitude of positive predictions made by the model, delineating the ratio of true positive predictions to the combined count of true positive and false positive predictions.
- #### **F1 Score** synthesizes Precision and Recall into a harmonized metric through their harmonic mean, offering a singular measure to appraise the model's balanced performance, particularly vital in contexts with asymmetric class distributions or when the penalty for false negatives significantly eclipses that for false positives.

### Advanced Confusion Matrix Visualization with Quantitative and Proportional Metrics for Model Diagnostics:

```python
def confusion_matrix_sklearn(model, predictors, target): # Create confusion matrix model with sklearn, predictors and target
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors) # predicting using the independent variables
    cm = confusion_matrix(target, y_pred) # creating a confusion matrix
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())] # percentage of each class
            for item in cm.flatten() # flatten confusion matrix
        ]
    ).reshape(2, 2) # reshape confusion matrix

    plt.figure(figsize=(6, 4)) # Constructing the demension of confusion matrix
    sns.heatmap(cm, annot=labels, fmt="") # plot confusion matrix heatmap with annotations
    plt.ylabel("True label") # Labeling y axis with "True label"
    plt.xlabel("Predicted label") # Labeling x axis with "Predicted label"
```

### Observations:
The above confusion_matrix_sklearn function enhances the analytical rigor of interpreting confusion matrices in classification model evaluations. It automates the generation of predictions using a specified classifier against a set of predictors and the target variable, thereafter computing the confusion matrix to capture the distribution of prediction outcomes across actual classes.

The above function's innovation lies in its procedural augmentation of the raw confusion matrix data, embedding both absolute frequencies and their proportional equivalents within each matrix cell. It achieves this through a sophisticated array transformation process, converting each element of the confusion matrix into a formatted string that encapsulates the count and its percentage of the aggregate sum of observations. This dual-layer annotation provides an immediate, intuitive grasp of the classifier’s performance dynamics, accentuating the comparative analysis of True Positive, False Positive, True Negative, and False Negative rates.

### Optimization of Decision Tree Classifier Parameters via Grid Search:

```python
# Choose the type of classifier.
estimator = DecisionTreeClassifier(random_state=1) # Create an instance of the DecisionTreeClassifier class with random_state=1

# Grid of parameters to choose from
parameters = { # Create a dictionary of parameters to pass to the grid search
    "max_depth": np.arange(6, 15), # Maximum depth of the tree
    "min_samples_leaf": [1, 2, 5, 7, 10], # Minimum number of samples required to be at a leaf node
    "max_leaf_nodes": [2, 3, 5, 10], # Maximum number of leaf nodes
}

# Type of scoring used to compare parameter combinations
acc_scorer = make_scorer(recall_score) # Create an instance of the make_scorer function

# Run the grid search
grid_obj = GridSearchCV(estimator, parameters, scoring=acc_scorer, cv=5) # Create an instance of the GridSearchCV class
grid_obj = grid_obj.fit(X_train, y_train) ## Complete the code to fit model on train data

# Set the clf to the best combination of parameters
estimator = grid_obj.best_estimator_ # Set the estimator to the best combination of parameters

# Fit the best algorithm to the data.
estimator.fit(X_train, y_train) ## Complete the code to fit model on train data
```
